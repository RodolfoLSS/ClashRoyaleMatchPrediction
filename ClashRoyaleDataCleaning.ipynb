# Is it possible to predict the winner of a Clash Royale match based only on the players' decks?

```python
import pandas as pd
import glob

path = r'/Users/RodolfoSaldanha/Downloads/CR_data' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

battles = pd.concat(li, axis=0, ignore_index=True)
battles.head()
```

### Select only battles 1vs1 on the Ladder mode

```python
battles.drop_duplicates(inplace=True)
```

```python
dataset = battles[battles['type']=='PvP']
dataset
```

# It's necessary to create a column describing who's the winner of the match, which will be the target of the classification.

1 = Victory of player 1

0 = Draw

-1 = Victory of player 2

```python
dataset['result'] = dataset['team_0_crowns'] - dataset.opponent_0_crowns
```

```python
dataset['result'].loc[dataset['result']>1] = 1
dataset['result'].loc[dataset['result']<-1] = -1
```
